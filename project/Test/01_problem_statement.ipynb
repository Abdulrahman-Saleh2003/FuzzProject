## 1. Problem Description

The Syrian consumer faces a complex challenge: managing multiple energy sources (solar panels, batteries, state grid) under a steep two-tier tariff system.

**Syrian Ministry of Electricity Tariff:**
- Subsidised tier: ≤ 500 kWh/month → **6 SYP / kWh**
- Unsubsidised tier: > 500 kWh/month → **18 SYP / kWh** (3× more expensive)

**SEMTOS** (Smart Energy Management & Tariff Optimization System) is a Fuzzy Expert System that automates real-time decisions to:
1. **Maximise economic efficiency** — stay within the subsidised 500 kWh tier.
2. **Protect hardware longevity** — manage battery depth-of-discharge and prevent overcharge.


## 2. Fuzzy Modeling Process

### Step 1: Define Variables

**Inputs (5):**
- `soc`      → Battery State of Charge, range [0, 100]  — labels: low, medium, full
- `solar`    → Solar panel output (W),  range [0, 5000] — labels: none, moderate, peak
- `grid`     → Grid availability,       range [0, 1]    — labels: cut, unstable, stable
- `cum_kwh`  → Cumulative monthly consumption (kWh), range [0, 1000] — labels: safe, critical, exceeded
- `demand`   → Current load demand (W), range [0, 10000] — labels: light, medium, heavy

**Outputs (3):**
- `bat_disch`  → Battery discharge rate, range [0, 100] — labels: save, eco, full
- `grid_use`   → Grid dependency level,  range [0, 100] — labels: bypass, loads_only, loads_plus_charge
- `load_ctrl`  → Load control mode,      range [0, 100] — labels: critical_only, eco_mode, full_operation


```
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl

# --- Input universes ---
soc      = ctrl.Antecedent(np.arange(0,   101, 1), 'soc')
solar    = ctrl.Antecedent(np.arange(0,  5001, 1), 'solar')
grid     = ctrl.Antecedent(np.arange(0,  1.01, 0.01), 'grid')
cum_kwh  = ctrl.Antecedent(np.arange(0,  1001, 1), 'cum_kwh')
demand   = ctrl.Antecedent(np.arange(0, 10001, 1), 'demand')

# --- Output universes ---
bat_disch  = ctrl.Consequent(np.arange(0, 101, 1), 'bat_disch')
grid_use   = ctrl.Consequent(np.arange(0, 101, 1), 'grid_use')
load_ctrl  = ctrl.Consequent(np.arange(0, 101, 1), 'load_ctrl')
```

### Step 2: Define Membership Functions

```
# --- SoC (%) ---
soc['low']    = fuzz.trapmf(soc.universe,    [0,   0,  20,  40])
soc['medium'] = fuzz.trapmf(soc.universe,    [30,  45,  55,  70])
soc['full']   = fuzz.trapmf(soc.universe,    [60,  80, 100, 100])

# --- Solar output (W) ---
solar['none']     = fuzz.trapmf(solar.universe, [0,    0,  500, 1000])
solar['moderate'] = fuzz.trapmf(solar.universe, [800, 1500, 3000, 3500])
solar['peak']     = fuzz.trapmf(solar.universe, [3000, 4000, 5000, 5000])

# --- Grid availability [0-1] ---
grid['cut']      = fuzz.trapmf(grid.universe, [0,    0,   0.2, 0.3])
grid['unstable'] = fuzz.trapmf(grid.universe, [0.2,  0.35, 0.55, 0.7])
grid['stable']   = fuzz.trapmf(grid.universe, [0.6,  0.75, 1.0,  1.0])

# --- Cumulative kWh ---
cum_kwh['safe']     = fuzz.trapmf(cum_kwh.universe, [0,   0,   350, 400])
cum_kwh['critical'] = fuzz.trapmf(cum_kwh.universe, [350, 430, 520, 600])
cum_kwh['exceeded'] = fuzz.trapmf(cum_kwh.universe, [550, 650, 1000, 1000])

# --- Current demand (W) ---
demand['light']  = fuzz.trapmf(demand.universe, [0,    0,  1500, 2500])
demand['medium'] = fuzz.trapmf(demand.universe, [2000, 3000, 5000, 6000])
demand['heavy']  = fuzz.trapmf(demand.universe, [5000, 6500, 10000, 10000])

# --- Battery discharge rate ---
bat_disch['save'] = fuzz.trapmf(bat_disch.universe, [0,  0,  10, 30])
bat_disch['eco']  = fuzz.trapmf(bat_disch.universe, [25, 40,  60, 70])
bat_disch['full'] = fuzz.trapmf(bat_disch.universe, [65, 80, 100, 100])

# --- Grid dependency level ---
grid_use['bypass']           = fuzz.trapmf(grid_use.universe, [0,  0,  10, 30])
grid_use['loads_only']       = fuzz.trapmf(grid_use.universe, [25, 40,  60, 65])
grid_use['loads_plus_charge']= fuzz.trapmf(grid_use.universe, [60, 75, 100, 100])

# --- Load control mode ---
load_ctrl['critical_only']   = fuzz.trapmf(load_ctrl.universe, [0,  0,  10, 30])
load_ctrl['eco_mode']        = fuzz.trapmf(load_ctrl.universe, [25, 40,  60, 65])
load_ctrl['full_operation']  = fuzz.trapmf(load_ctrl.universe, [60, 75, 100, 100])
```

## 4. Rule Base (If-Then Rules)

Five core rules covering the most common Syrian energy scenarios:

| # | Rule | Rationale |
|---|------|-----------|
| R1 | If grid=stable AND cum_kwh=exceeded AND solar=peak → bypass grid, eco discharge, eco load | Tariff escape: grid is 18 SYP — use sun+battery |
| R2 | If grid=stable AND cum_kwh=safe AND soc=low → loads+charge, save battery, full load | Cheap tier (6 SYP) — charge battery while running everything |
| R3 | If grid=cut AND solar=none AND soc=low → save battery, critical load only | Emergency: prevent deep discharge |
| R4 | If solar=peak AND soc=full AND demand=heavy → bypass grid, full discharge, full load | 100% solar self-sufficiency |
| R5 | If grid=unstable AND cum_kwh=critical AND soc=medium → loads_only, eco load | Caution mode near 500 kWh boundary |


```
# R1 — Tariff escape: grid costs 18 SYP, switch to solar+battery
rule1 = ctrl.Rule(
    grid['stable'] & cum_kwh['exceeded'] & solar['peak'],
    (bat_disch['eco'], grid_use['bypass'], load_ctrl['eco_mode'])
)

# R2 — Exploit cheap subsidised tier: charge battery + run full house
rule2 = ctrl.Rule(
    grid['stable'] & cum_kwh['safe'] & soc['low'],
    (bat_disch['save'], grid_use['loads_plus_charge'], load_ctrl['full_operation'])
)

# R3 — Emergency: grid cut, no sun, battery critical → protect battery
rule3 = ctrl.Rule(
    grid['cut'] & solar['none'] & soc['low'],
    (bat_disch['save'], load_ctrl['critical_only'])
)

# R4 — Peak solar + full battery + heavy demand → 100% self-sufficient
rule4 = ctrl.Rule(
    solar['peak'] & soc['full'] & demand['heavy'],
    (bat_disch['full'], grid_use['bypass'], load_ctrl['full_operation'])
)

# R5 — Near 500 kWh boundary + unstable grid → caution mode
rule5 = ctrl.Rule(
    grid['unstable'] & cum_kwh['critical'] & soc['medium'],
    (grid_use['loads_only'], load_ctrl['eco_mode'])
)
```

## 5. Build and Simulate Control System
